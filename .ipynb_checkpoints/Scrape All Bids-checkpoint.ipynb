```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import pickle
import os
import numpy as np
```

```python
urlbase = r'http://ff.bluebones.net/auction-results.php?bid_round_id='
```

```python
dicturls = {'div_1_urls' : [1],
'div_2_urls' : [36],
'div_3a_urls' : [32],
'div_3b_urls' : [30],
'div_3c_urls' : [26],
'div_4a_urls' : [25],
'div_4b_urls' : [27],
'div_4c_urls' : [34],
'div_4d_urls' : [29],
'div_5a_urls' : [24],
'div_5b_urls' : [28],
'div_5c_urls' : [31],
'div_5d_urls' : [33],
'div_5e_urls' : [35],
'div_5f_urls' : [14],
'div_6a_urls' : [23],
'div_6b_urls' : [3],
'div_6c_urls' : [4],
'div_6d_urls' : [5],
'div_6e_urls' : [6],
'div_6f_urls' : [7],
'div_6g_urls' : [8],
'div_6h_urls' : [9],
'div_6i_urls' : [10],
'div_7a_urls' : [11],
'div_7b_urls' : [12],
'div_7c_urls' : [2],
'div_7d_urls' : [13],
'div_7e_urls' : [15],
'div_7f_urls' : [16],
'div_7g_urls' : [17],
'div_7h_urls' : [18],
'div_7i_urls' : [19],
'div_7j_urls' : [20],
'div_7k_urls' : [21],
'div_7l_urls' : [22],
'div_7m_urls' : [37]}
```

```python
urllist = list(dicturls.values())
```

```python
url_numbers = np.empty((len(urllist),20))
url_numbers[:] = np.NaN
```

```python
a = 0
for l in urllist:
    curl = urlbase + str(l[0])
    ra = requests.get(curl)
    soupa = BeautifulSoup(ra.content, "lxml")
    rounds_rough = soupa.find_all('a')[250:275]
    rounds = [c['href'] for c in rounds_rough if "Round" in c.text]
    round_ids = [l[0]] + [int(c.split("=")[1]) for c in rounds]
    lrid = len(round_ids)
    url_numbers[a,:lrid] = np.asarray(round_ids)
    a += 1
```

```python
url_df = pd.DataFrame(data=url_numbers.T, columns=list(dicturls.keys())).dropna(how='all')
```

```python
boolmap = {'No' : False, 'Yes' : True}
column_names = ["Division","Level","Round","Player","Manager","Team","Position","Amount","Success"]
df_allbids = pd.DataFrame(columns = column_names)
```

```python
for c in range(url_df.shape[1]):
    print(c)
    div = url_df.iloc[:,c].name.split('_')[1]
    level = div[0]
    for r in range(url_df.iloc[:,c].dropna().size):
        url = urlbase + str(int(url_df.iloc[r,c]))
        ra = requests.get(url)
        soupa = BeautifulSoup(ra.content, "lxml")
        tables = soupa.find_all('table')
        if len(tables) > 0:
            table = tables[0]
            tr = table.find_all('tr')
            for t in range(1,len(tr)):
                entry = tr[t].text.split('\n')
                player = entry[1]
                manager = entry[2]
                team = entry[3]
                position = entry[4]
                amount = float(entry[5][1:-1])
                success = boolmap[entry[6]]
                df_allbids = df_allbids.append({"Division":div,"Level":level,"Round":(r+1),"Player":player,"Manager":manager,"Team":team,"Position":position,"Amount":amount,"Success":success},ignore_index=True)
```
