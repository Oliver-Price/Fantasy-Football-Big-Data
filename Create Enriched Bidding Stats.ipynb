```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from IPython.core.display import HTML
import datetime
import numpy as np
import numpy.ma as ma
import os
```

```python
#year = 18; w = 470
year = 19; w = 543
```

```python
save = os.getcwd()
save = os.path.join(save,"save")
save = os.path.join(save,str(year))
save = os.path.join(save,'df.pickle')
    
with open(save, 'rb') as f:
    df = pickle.load(f)
```

  ## Add basic extra columns

```python
df2 = df.filter(regex='Price').fillna(0)
df['Average Cost'] = df2.mean(axis = 1,skipna=True)
df['Max Cost'] = df2.max(axis = 1,skipna=True)
df['Cost Deviation'] = df2.std(axis = 1,skipna=True)
df['Zero Bids'] = df2.apply( lambda s : s.value_counts().get(0,0), axis=1)
aves = df.mean(axis = 0, skipna = True)
mcpp = aves['Max Cost']/aves['Points']
df['Rough Guide'] = mcpp*df['Points']
df['Underrated'] = df['Rough Guide'] - df['Max Cost']
```

## Add Round by Round info

```python
df2 = df.filter(regex='Price')
df3 = df.filter(regex='When')
```

```python
#2018
r1 = datetime.datetime(2018, 8, 2)
r2 = datetime.datetime(2018, 8, 5)
r3 = datetime.datetime(2018, 8, 6)
r4 = datetime.datetime(2018, 8, 7)
```

```python
#2019
r1 = datetime.datetime(2019, 7, 31)
r2 = datetime.datetime(2019, 8, 3)
r3 = datetime.datetime(2019, 8, 4)
r4 = datetime.datetime(2019, 8, 5)
```

```python
r1bool = df3 == pd.Timestamp(r1)
r1s = np.nan_to_num(r1bool.values*df2.values)
r1only = ma.array(df2.values, mask = ~r1bool.values)
df['Round 1 Average'] = ma.mean(r1only,axis=1)
df['Round 1 Bids'] = r1bool.sum(axis=1)
maxr1 = np.max(r1s,axis=1)
df['Highest R1 Bid'] = maxr1
df['Round 1 Activty'] = np.any(r1bool.values,axis=1)
```

```python
r2bool = df3 == pd.Timestamp(r2)
r2s = np.nan_to_num(r2bool.values*df2.values)
r2only = ma.array(df2.values, mask = ~r2bool.values)
df['Round 2 Average'] = ma.mean(r2only,axis=1)
df['Round 2 Bids'] = r2bool.sum(axis=1)
maxr2 = np.max(r2s,axis=1)
df['Highest R2 Bid'] = maxr2
df['Round 2 Activty'] = np.any(r2bool.values,axis=1)
```

```python
r3bool = df3 == pd.Timestamp(r3)
r3s = np.nan_to_num(r3bool.values*df2.values)
r3only = ma.array(df2.values, mask = ~r3bool.values)
df['Round 3 Average'] = ma.mean(r3only,axis=1)
df['Round 3 Bids'] = r3bool.sum(axis=1)
maxr3 = np.max(r3s,axis=1)
df['Highest R3 Bid'] = maxr3
df['Round 3 Activty'] = np.any(r3bool.values,axis=1)
```

```python
r4bool = df3 == pd.Timestamp(r4)
r4s = np.nan_to_num(r4bool.values*df2.values)
r4only = ma.array(df2.values, mask = ~r4bool.values)
df['Round 4 Average'] = ma.mean(r4only,axis=1)
df['Round 4 Bids'] = r4bool.sum(axis=1)
maxr4 = np.max(r4s,axis=1)
df['Highest R4 Bid'] = maxr4
df['Round 4 Activty'] = np.any(r4bool.values,axis=1)
```

## Save enriched dataset

```python
save = os.getcwd()
save = os.path.join(save,"save")
save = os.path.join(save,str(year))
save = os.path.join(save,'df_rich.pickle')

with open (save, 'wb') as f:
    pickle.dump(df, f)
```
